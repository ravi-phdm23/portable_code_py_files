# File Extension Renamer

This notebook renames file extensions in the backend folder and can revert them.

```python
import os
```

```python
def rename_extensions(directory, mode):
    extensions = ['.py', '.txt', '.json', '.html', '.css', '.js', '.map']
    
    if mode == 'rename':
        for root, dirs, files in os.walk(directory):
            for file in files:
                for ext in extensions:
                    if file.endswith(ext):
                        old_path = os.path.join(root, file)
                        new_file = file.replace(ext, '_' + ext[1:])
                        new_path = os.path.join(root, new_file)
                        if os.path.exists(new_path):
                            print(f"Skipping {old_path}: {new_path} already exists")
                        else:
                            os.rename(old_path, new_path)
                            print(f"Renamed {old_path} to {new_path}")
                        break
    elif mode == 'revert':
        for root, dirs, files in os.walk(directory):
            for file in files:
                for ext in extensions:
                    suffix = '_' + ext[1:]
                    if file.endswith(suffix):
                        old_path = os.path.join(root, file)
                        new_file = file.replace(suffix, ext)
                        new_path = os.path.join(root, new_file)
                        if os.path.exists(new_path):
                            print(f"Skipping {old_path}: {new_path} already exists")
                        else:
                            os.rename(old_path, new_path)
                            print(f"Renamed {old_path} to {new_path}")
                        break
    else:
        print("Invalid mode. Use 'rename' or 'revert'.")
```

## Rename Extensions

```python
rename_extensions('.', 'rename')
```

## Revert Extensions

```python
rename_extensions('.', 'revert')
```
